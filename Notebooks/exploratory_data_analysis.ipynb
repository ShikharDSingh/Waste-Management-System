```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('cleaned_data.csv')
```

```python
df.head()
```

```python
df['Recycling Rate (%)'].describe()
```

```python
# Separate categorical and numeric data
categorical_cols = df.select_dtypes(include=['object']).columns
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Categorical data
categorical_data = df[categorical_cols]

# Numeric data
numeric_data = df[numeric_cols]
```

```python
categorical_data_cleaned = categorical_data.dropna()
numeric_data_cleaned = numeric_data.dropna()
```

```python
# Distribution of Recycling Rate
sns.histplot(df["Recycling Rate (%)"], kde=True)
plt.title("Distribution of Recycling Rate (%)")
plt.show()
```

```python
# Recycling Rate by Waste Type
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x="Waste Type", y="Recycling Rate (%)")
plt.xticks(rotation=45)
plt.title("Recycling Rate by Waste Type")
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
